# Extract CoT to Instructions

```python
import pandas as pd
```

```python
df = pd.read_csv("./results/v2/steerbench_v2.1_llama3.1_cot.csv", index_col=0)
df["raw_response"]
```

```python
import re

def get_cot_block(text):
    pattern = r"## Edits(.*?)## Rewritten text"
    match = re.search(pattern, text, re.DOTALL)
    if match:
        content = match.group(1).strip()
    else:
        content = None
    return content

df["raw_response"].apply(get_cot_block).tolist()
```
