```python
import pandas as pd

df = pd.read_csv("./training_probes/steerability_2d_in_3d_64x16_double_trouble_ipo_beta_eval.csv")
df.head()
```

```python
import ast
import requests
import numpy as np

def print_prompt(i):
    print(ast.literal_eval(df.iloc[i]["prompt"])[0]["content"])

def get_source_target(i):
    temp = pd.DataFrame(
        np.stack([df.iloc[i].filter(regex="^source_").values, df.iloc[i].filter(regex="^target").values], axis=1),
        index=df.filter(regex="^target").columns.str[len("target_"):],
        columns=["source", "target"]
    )
    return temp

def map_to_goalspace(s: str, port=12121):
    session = requests.Session()
    resp = session.post(f"http://127.0.0.1:{port}/goalspace", json={"texts": [s]})
    if resp.status_code == 200:
        return resp.json()
    else:
        raise ValueError(f"HTTP Error: {resp.status_code}")
    
get_source_target(0)
```
