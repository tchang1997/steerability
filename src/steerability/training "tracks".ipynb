```python
training_probe = pd.read_csv("./training_probes/llama3.1_micro_steer_grpo_1in2d_n1_i16_cce_lora_v2_eval.csv", index_col=0)
training_probe
```

```python
import matplotlib.pyplot as plt

n = 0

completions = df.groupby("step").nth(n)["completion"]
mappings = await map_to_goalspace(completions.tolist())
example_df = pd.DataFrame(mappings)
```

```python
import matplotlib.patches as patches
import numpy as np

# Assuming the DataFrame has columns 'x' and 'y'
x = example_df["sadness"].values
y = example_df["surprise"].values

source_sad =  training_probe.iloc[n]["source_sadness"] 
source_surp =  training_probe.iloc[n]["source_sadness"] 

target_sad = training_probe.iloc[n]["target_sadness"] 
target_surp = training_probe.iloc[n]["target_surprise"]

```

```python

# Create a colormap based on time ordering (most recent = red)
colors = np.linspace(0, 1, len(example_df))  # Normalize time for colormap
cmap = plt.cm.jet

fig, ax = plt.subplots(figsize=(4, 3))

# Scatter plot of points with color
sc = ax.scatter(x, y, c=colors, cmap=cmap, marker=".", s=20, alpha=np.linspace(0, 1, len(example_df)))

# Add arrows between points (using light-colored arrows)
for i in range(len(x)):
    ax.annotate(
        "",
        xy=(x[i], y[i]), xytext=(source_sad, source_surp), #(x[i], y[i]),
        arrowprops=dict(arrowstyle="->", color=cmap(colors[i]), alpha=colors[i], lw=1)
    )

tol = 0.2

# Define boundaries
x_min, x_max = 0, 1
y_min, y_max = 0, 1

# Calculate rectangle boundaries ensuring it stays within [0, 1]
rect_x = max(target_sad - tol, x_min)
rect_y = max(target_surp - tol, y_min)
rect_width = min(2*tol, x_max - rect_x)
rect_height = min(2*tol, y_max - rect_y)

rect = patches.Rectangle(
    (rect_x, rect_y),  # Bottom-left corner
    rect_width,  # Width (0.2 left + 0.2 right)
    rect_height,  # Height (0.2 down + 0.2 up)
    color='red', alpha=0.1  # Adjust color & transparency
)
ax.add_patch(rect)
ax.annotate("", xy=(target_sad, target_surp),
            xytext=(source_sad, source_surp), arrowprops=dict(arrowstyle="->", color="black", alpha=0.8, lw=2))
    
# Invert the colormap so red is the latest
sc.set_cmap(plt.cm.jet)

# Colorbar for reference
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("Epoch (prop. of training)")

# Labels and title
ax.set_xlabel("Sadness")
ax.set_ylabel("Surprise")
ax.set_xlim((0, 1))
ax.set_ylim((0, 1))
ax.set_title(f"Instruction #{n}, optimization over time")

plt.grid(True)
```
